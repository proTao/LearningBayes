**问题描述**
放射源以平均每秒r个粒子的速度向一个Geiger计数器发射粒子，但是计数器的计数成功率是f。若计数器一秒的时间记录了15个粒子，那么实际击中计数器的粒子数n的后验分布是？粒子平均发射速率r的后验分布是？

**建模思路**
n是以r为参数的泊松分布。
记录的粒子数是以(n,f)为参数的二项分布。

这是一个两层模型r -> n -> 15


**单层假设**
先假设r已知，是一个常数

```python
import thinkplot
from thinkbayes2 import *
import numpy as np
from scipy import stats
from matplotlib import pyplot as plt
from operator import __lt__, __le__, __gt__, __ge__
```

```python
class Hit(Suite):
    # 对实际击中的粒子数建模，被r控制的泊松分布
    def __init__(self, r, f=0.1, high=200, label="no label"):
        self.r = r
        self.f = f
        Suite.__init__(self, MakePoissonPmf(high=high, step=int(high/80), lam=r), label=label)
    
    def Likelihood(self, data, hypo):
        # data是显示出的粒子数
        # hypo是击中的粒子数
        like = stats.binom.pmf(k=data, n=hypo, p=self.f)
        return like

```

```python
hit_100 = Hit(100, high=500, label="100")
hit_200 = Hit(200, high=500, label="200")
hit_400 = Hit(400, high=500, label="400")
thinkplot.Pmfs([hit_100, hit_200, hit_400])

thinkplot.Config(legend=True)
plt.show()
```

**如下图所示**
固定n为被参数100,200,400控制的泊松分布时，观察到15对泊松先验的影响

```python
hit_100.Update(15)
hit_200.Update(15)
hit_400.Update(15)
thinkplot.Pmfs([hit_100, hit_200, hit_400])

thinkplot.Config(legend=True)
plt.show()
```

接下来就是使先验不是固定参数下的分布，这又是一次进步

- 在频率学派下，我们固定先验是一个未知常数，数据在该常数的控制下分布中得到，得到数据后通过极大似然去估计该常数
- 在单层贝叶斯估计下，我们令先验是一个固定常数控制的分布，数据的生成过程是该先验分布中的一个随机常数的控制下的某种随机试验，得到数据后更新该先验分布得到后验分布
- 在双层贝叶斯估计下，我们令先验是一个非固定常数控制的分布。但是其实就是使先验知识往前推了一层，即我们的先验知识不是参数的分布，是参数的控制参数的分布。

**多层建模**
接下来对更前一层建模，前面我们假设击中粒子数是固定常数下控制的泊松分布，那么这一层我们对泊松参数lambda建模。
认为lambda是一个正态分布。

```python
class Emitter(Suite):
    def __init__(self, mu, sigma, samples=30):
        lam_dist = MakeNormalPmf(mu=mu, sigma=sigma, num_sigmas=2, n=samples)
        delete = []
        for r in lam_dist:
            if r <= 0:
                delete.append(r)
        for r in delete:
            lam_dist.Remove(r)
        Suite.__init__(self,{Hit(r):prob for r, prob in lam_dist.Items()})
        
    def Likelihood(self, data, hypo):
        # data是记录下来的粒子数
        # hypo是一个常数r控制下的击中数分布
        like = hypo.Update(data)
        return like
    
    def DistOfR(self):
        items = [(hit_dist.r, prob)for hit_dist, prob in self.Items()]
        return MakePmfFromItems(items)
    
    def DistOfN(self):
        return MakeMixture(self)
        
```

```python
radion = Emitter(mu=120,sigma=10)
r0 = radion.DistOfR()
r0.label="init"
thinkplot.Pmf(r0)

radion.Update(15)
r1 = radion.DistOfR()
r1.label="update 1"
thinkplot.Pmf(r1)

radion.Update(15)
r2 = radion.DistOfR()
r2.label="update 2"
thinkplot.Pmf(r2)

thinkplot.Config(legend=True)
plt.show()
```

```python
radion = Emitter(mu=120,sigma=10)
r0 = radion.DistOfN()
r0.label="init"
thinkplot.Pmf(r0)

radion.Update(15)
r1 = radion.DistOfN()
r1.label="update 1"
thinkplot.Pmf(r1)

radion.Update(15)
r2 = radion.DistOfN()
r2.label="update 2"
thinkplot.Pmf(r2)

thinkplot.Config(legend=True)
plt.show()
```
